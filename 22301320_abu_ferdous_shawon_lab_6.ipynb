```python
#1(a)
#Factorial
def factorial(num):
  if num == 1:
    return 1
  else:
    return num * (factorial(num-1))

factorial(2)
```

```python
#1(b)
#Fibonacci
def Fibonacci(num):
  if num<2:
    return num
  return Fibonacci(num-1)+Fibonacci(num-2)

Fibonacci(11)
```

```python
#1 c) Print all the elements of a given array recursively.
def printA(arr,idx):
  if idx != len(arr):
    print(arr[idx])
    printA(arr,idx+1)

printA([19,23,52,24,65],0)
```

```python
#1(d) power
def powerN(base,n):
  if n==1:
    return base
  else:
    return base*(powerN(base,n-1))

powerN(3,3)
```

```python
#2(a) Binary
def binary(n):
  if n == 0 or n==1:
    return str(n)
  else:
    return binary(n // 2) + str(n % 2)

binary(5)
```

```python
# 2 b) Implement a recursive algorithm to add all the elements of a non-dummy
# headed singly linked linear list. Only head of the list will be given as
# parameter where you may assume every node can contain only integer
# as its element.
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next
def sum(head):
  if head.next == None:
    return 0
  else:
    return head.elem + sum(head.next)
head = [15,10,34,41,56,72]
sum(head)
```

```python
# 2 c) Implement a recursive algorithm which will print all the elements of a non-dummy
# headed singly linked linear list in reversed order.
def reverse_print(head):
  if head == None:
    return
  else:
    reverse_print(head.next)
    print(head.elem)

head = [15,10,34,41,56,72]
reverse_print(head)
```

```python
#3
def hocBuilder(height):
  if height == 0:
    print("House cannot be build")
  elif height == 1:
    return 8
  else:
    return 5 + hocBuilder(height-1)

hocBuilder(2)
```

```python
#4 (a)
def new_loop(num):
  if num==0:
    return
  else:
    new_loop(num-1)
    print(num,end="")


def pattern(n):
  if n==0:
    return
  else:
    pattern(n-1)
    new_loop(n)
    print()


pattern(5)
```

```python
#4(b)
def space(n):
  if n==0:
    return
  else:
    space(n-1)
    print(" ",end="")

def print_elem(n):
  if n==0:
    return
  else:
    print_elem(n-1)
    print(n,end="")

def pattern(n,i=1):
  if n==0:
    return
  else:
    space(n-1)
    print_elem(i)
    print()
    pattern(n-1,i=i+1)

pattern(5)
```
