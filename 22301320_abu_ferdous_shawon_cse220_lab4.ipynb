```python
class Node:
  def __init__(self,elem,next,prev):
    self.elem=elem
    self.next=next
    self.prev=prev
class Patient:
  def __init__(self,id,name,age,bg):
    self.id=id
    self.name=name
    self.age=age
    self.bg=bg
class WRM:
  def __init__(self):
    self.head=Node(None,None,None)
    self.head.next=self.head
    self.head.prev=self.head
    self.count=None

  def RegisterPatient(self):
    if self.count==None:
      self.count=0
    else:
      self.count+=1
    p1= Patient(int(input("Please enter patient's ID: ")),input("Please enter patient's Name: "),int(input("Please enter patient's age: ")),input("Please enter patient's Blood Group: "))
    n=Node(p1,None,None)
    if self.count==0:
      tail=self.head
      tail.next=n
      tail=n
      tail.prev=self.head
      tail.next=self.head
      self.head.prev=tail
      print("SUCESSFULLY REGISTERED",p1.name)
    else:
      tail = self.head.next
      while tail!=self.head:
        tail = tail.next
      tail.prev.next = n
      tail.prev.next.prev = tail.prev
      tail.prev.next.next = tail
      tail.prev = tail.prev.next
      print("SUCESSFULLY REGISTERED",p1.name)
  def ServePatient(self):
    if self.head.next==self.head:
      print("No patient left")
    else:
      print(f"SERVING {self.head.next.elem.name}")
      self.head.next = self.head.next.next
      self.head.next.prev = self.head

  def CancelAll(self):
    self.head.next = self.head
    self.head.prev = self.head
    print("ALL APPOINMENT HAS BEEN CANCELLED")

  def CanDoctorGoHome(self):
    if self.head.next==self.head:
      print("YES, HE CAN GO TO HOME")
    else:
      print("NO, MORE PATIENTS ARE WAITING")
  def ShowAllPatient(self):
    tail=self.head.next
    serial=1
    while tail!=self.head:
      print(f"{serial}. {tail.elem.id}")
      serial+=1
      tail=tail.next

print("WELCOME TO BRACU HOSPITAL")
a=WRM()
while True:
  x = int(input("press 1 to register\npress 2 to serve patient\npress 3 to cancel all appointments\npress 4 to know if doctor can go home or not\npress 5 to see all patient\npress 6 to exit\n"))
  if x==1:
    a.RegisterPatient()
  elif x==2:
    a.ServePatient()
  elif x==3:
    a.CancelAll()
  elif x==4:
    a.CanDoctorGoHome()
  elif x==5:
    a.ShowAllPatient()
  elif x==6:
    break

print("THANKS FOR TAKING OUR SERVICE. COME AGAIN!")
```
